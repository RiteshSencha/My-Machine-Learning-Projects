```python
import numpy as np
x = np.arange(0,100) 
y = x*2
z = x**2
```


# Question 1

```python
plt.plot(x, y, 'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Question 1 solution')
plt.show()
```

# Question 3

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.1, 0.6, 0.4, 0.3]) # inset axes
# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Axes 1 Title')
# Insert Figure Axes 2
axes2.plot(y, x, 'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Axes 2 Title');
```
