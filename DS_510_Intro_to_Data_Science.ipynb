# Import Libraries & Data

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
HR = pd.read_csv('Final HRA Data.csv')
print(HR)
```

```python
#Lets take a look at data types of all columns
HR.dtypes
```

# We will now check what our data looks like in a table format with only first 10 rows

```python
HR.head(10)
```

# To see count of any given attribute, we will use the below code and change it to whatever column we are trying to look at

```python
HR['Department'].value_counts().sort_index().head(10)
```

#Data Visualization

```python
HR['EducationField'].value_counts().sort_index().plot(kind='pie')
```

```python
HR['YearsInCurrentRole'].value_counts().sort_index().plot(kind='bar')
```

```python
HR['WorkLifeBalance'].value_counts().sort_index().plot(kind='bar')
```

```python
HR['JobSatisfaction'].value_counts().sort_index().plot(kind='bar')
```

```python
HR['NumCompaniesWorked'].value_counts().sort_index().plot();
```

# Making histograms of entire data set at once

```python
HR.hist(figsize=(25,15));
```

```python
HR.plot(kind='box', x='MonthlyIncome', y='WorkLifeBalance', \
        color='dodgerblue', figsize=(10,5))
```

```python
#reference for above
HR.plot(kind='scatter', x='Gender', y='JobSatisfaction', \
        color='dodgerblue', figsize=(10,4), s=10, alpha=0.5)
```
